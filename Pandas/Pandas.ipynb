### The groupby() function 
#### In pandas is used to group data based on one or more columns and then apply an aggregation function to each group. Here's an example to explain how groupby() works.

#### Example: Grouping Data by a Column and Aggregating
Let's say you have a DataFrame with information about sales for different products across various regions.

```python
import pandas as pd

# Sample DataFrame
data = {
    'Region': ['North', 'South', 'North', 'East', 'South', 'East', 'North'],
    'Product': ['A', 'B', 'A', 'C', 'A', 'B', 'C'],
    'Sales': [100, 150, 200, 130, 120, 90, 300]
}

df = pd.DataFrame(data)

df
```

```python
# Group by 'Region' and calculate the sum of 'Sales' for each region
grouped_sales = df.groupby('Region')['Sales'].sum()

print(grouped_sales)
```

##### Explanation : 
The groupby('Region') groups the data by the values in the 'Region' column.

['Sales'].sum() then sums the 'Sales' values for each group (region).

#### Example: Group by Multiple Columns and Use Multiple Aggregations
You can also group by multiple columns and apply different aggregation functions.

```python
# Group by 'Region' and 'Product', then calculate the sum and mean of 'Sales'
grouped = df.groupby(['Region', 'Product'])
print(grouped)
print("\n")
print(grouped.first())
print("\n")


## applying aggregation on Sales col
res=grouped.agg({'Sales': ['sum', 'mean']})
res
```

#### Explanation:
groupby(['Region', 'Product']) groups the data first by 'Region' and then by 'Product'.

.agg({'Sales': ['sum', 'mean']}) applies both the sum and mean aggregation functions to the 'Sales' column for each group.

```python
import pandas as pd

# Sample DataFrame
data = {
    'Region': ['North', 'South', 'North', 'East', 'South', 'East', 'North'],
    'Product': ['A', 'B', 'A', 'C', 'A', 'B', 'C'],
    'Sales': [100, 150, 200, 130, 120, 90, 300],
    'Quantity': [10, 20, 30, 40, 50, 60, 70]
}

df = pd.DataFrame(data)

# Group by 'Region' and apply multiple aggregation functions on 'Sales' and 'Quantity'
grouped = df.groupby('Region').agg({
    'Sales': ['sum', 'mean', 'max'],
    'Quantity': ['sum', 'min']
})

print(grouped)
```

#### 7. Concatenating DataFrames
You can concatenate multiple DataFrames vertically or horizontally.

##### 1. Vertical Concatenation (Stacking Rows)
Vertical concatenation adds the rows of one DataFrame below the other, stacking them on top of each other. This is useful when you have DataFrames with the same columns.

```python
import pandas as pd

# Create two example DataFrames
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

print(df1)
print('\n')
print(df2)
print("\n")
# Concatenate vertically
result = pd.concat([df1, df2], axis=0, ignore_index=True)
print(result)
```

axis=0: Specifies vertical concatenation (by rows).

ignore_index=True: Resets the index for the concatenated DataFrame.

##### 2. Horizontal Concatenation (Merging Columns)
Horizontal concatenation adds the columns of one DataFrame to another, side by side. This is useful when you have DataFrames with the same index but different columns.

```python
df3 = pd.DataFrame({'A': [9, 10], 'C': [11, 12]})

# Concatenate vertically with different columns
result = pd.concat([df1, df3], axis=0, ignore_index=True)
print(result)
```

##### Key Points:
axis=0 for vertical concatenation (rows).

axis=1 for horizontal concatenation (columns).

ignore_index=True to reset the index after concatenation.

## Pandas joins

```python
#Inner (natural) joins
# importing pandas
import pandas as pd

# Creating dataframe a
a = pd.DataFrame()

# Creating Dictionary
d = {'id': [1, 2, 10, 12],
	'val1': ['a', 'b', 'c', 'd']}

a = pd.DataFrame(d)

# Creating dataframe b
b = pd.DataFrame()

# Creating dictionary
d = {'id': [1, 2, 9, 8],
	'val1': ['p', 'q', 'r', 's']}
b = pd.DataFrame(d)

# inner join
df = pd.merge(a, b, on='id', how='inner')

# display dataframe
df
```

```python
#left outer join
# left outer join
df1 = pd.merge(a, b, on='id', how='left')
  
# display dataframe
df1
```

```python
# right outer join
df2 = pd.merge(a, b, on='id', how='right')
  
# display dataframe
df2
```

```python
# full outer join
df3= pd.merge(a, b, on='id', how='outer')
  
# display dataframe
df3
```

##### Index Join: To merge the dataframe on indices pass the left_index and right_index arguments as True i.e. both the dataframes are merged on an index using default Inner Join.

```python
# index join
df4= pd.merge(a, b, left_index=True, right_index=True)
  
# display dataframe
df4
```
