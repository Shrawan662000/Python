# Matplotlib


```python
import matplotlib.pyplot as plt
x=[1,2,3,5,7]
y=[-5, 8, 4, 7, 3]
```

```python
plt.plot(x,y)
```

```python
fig=plt.figure(figsize=(10, 5)) #10 inch width and 5 inch height
plt.plot(x, y, color='violet')
```

```python
plt.plot(x,y)
plt.xlim(0,10)
plt.ylim(-5, 10)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title(" Line Plot")
plt.suptitle(" Data Visual", y=1.0)
```

```python
plt.savefig('data.png')
```
